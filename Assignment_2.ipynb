<b>Question 1</b> : What are semantic HTML elements? Why is using them important for web development? <br>
<b>Answer</b> : Semantic tags are HTML tags that give meaning to the content of a web page beyond just its presentation. <br>
They provide information about the structure and meaning of the content, making it easier for search engines, screen readers, and other tools to understand the page and its content. <br>
To understand semantics, i am building a web page using newly introduced structural-semantic elements in HTML. <br>
Let's code a web page, using all common semantic tags to define a common web page structure.

```python
<body>
    <!-- Header -->
     <header>
        <h1> Sementic HTML Tags </h1>
        <nav>
            <ul>
                <li><a href="#"> Home </a></li>
                <li><a href="#"> About Us </a></li>
                <li><a href="#"> Services</a></li>
                <li><a href="#"> Contact Us </a></li>
            </ul>
        </nav>
     </header>
     <!-- Main Content -->
      <main>
        <!-- Article-->
         <article>
            <h2> What are Sementic HTML Tags ?</h2>
            <p>Semantic HTML tags are tags that give meaning to the content they surround. They
provide information about the structure and meaning of the content on a web page, making it
easier for search engines, screen readers, and other tools to understand the page and its
content.</p>
         </article>
         <!-- Section-->
          <section>
            <h2>Sementic Tags</h2>
            <ul>
                <li>header : Indicates the beginning of a header section</li>
                <li>nav: Defines a section of navigation links</li>
                <li>article: Indicates an independent, self-contained piece of content</li>
                <li>footer: Defines a footer section at the end of a page or article</li>
            </ul>
          </section>
      </main>
      <!-- Footer-->
       <footer>
        <p>&copy; 2023 Example Company. All rights reserved.</p>
       </footer>
</body>
```

In the above example, we've used many semantic tags, such as < header >, < nav >, < main >, < article >, < footer >,
< h1 >, < p > to provide additional information about the structure and meaning of the content on the page.
These tags make the code more readable and easier to understand, both for humans and for web browsers
and other tools that might be parsing the HTML. <br> <br>

<b>Question 3: </b>How can you make an HTML form more accessible to users with disabilities? <br>
<b>Answer: </b> Accessibility in HTML refers to the practice of designing and coding web content in a way that ensures it can be
easily accessed, understood, and interacted with by people of diverse abilities, including those with disabilities.
This includes using semantic markup, providing alternative text for images and videos, and making sure your
web pages are navigable using a keyboard.
There are some assistive devices that play a major role in providing accessibility.<br>
Some best prectices are as follows: <br><br>
1)<b> Image alternative text - </b> Add descriptive alternative text (alt text) to all images using the alt attribute.<br>
<b>Example : </b> < img src="https://picsum.photos/200/300" alt="Random Pic"width="200" height="200" / > <br>
If the above code is not working due to connectivity issues or a mistake in the URL link or Path. The alternate text
will be displayed.<br>
2)<b> Ensuring proper heading structure or Semantic Structure of HTML- </b>practice the use of proper semantic structure, by using appropriate HTML elements like < h1 >, < nav >, < main >, < article >, < button >, and < label > to convey the purpose and role of each element.<br>
3)<b> Making links descriptive - </b>to make a link descriptive in HTML, meaning full and descriptive text in the anchor < a > element should be used.<br>
4)<b> Accessibility in Form -</b> use of < label > tag or aria-label attribute and Form fields should be properly grouped with  < fieldset > and < legend > elements.<br>
5)<b> Text resizes and Zoom testing -</b> The website pages should be checked by resizing text or zooming in/out to ensure that the content remains readable and usable without causing layout or functionality issues.<br>
6)<b> Video and Audio Accessibility - </b>Provide captions, transcripts, and audio descriptions for multimedia content. Ensure that video players are acces sible and controllable with the keyboard.<br>
7)<b> HTML tables - </b>Ensure that tables are used for tabular data and properly marked up with appropriate headers and scope attributes to provide context to screen readers.<br>
8)<b> Colour contrast - </b>Verify that the colour contrast between text and background meets the WCAG contrast requirements. Ensure that information conveyed through colour is also available via other means (e.g.,
labels, symbols).
9)<b> Tab-index -</b> The tab-index is an HTML attribute that can be used to specify the order of navigating an HTML element when the tab button is pressed. The tab-index can be used on any HTML element but it should be used only on specific elements where it is needed.

<b>Question 7:</b> How do the relative, absolute, and fixed positioning properties differ in CSS?<br>
<b>Answer : </b><br><b>CSS position property</b> allows you to control how an element is placed on the web page. By default, elements follow the normal document flow, but with the position property, you can change this behavior and place elements exactly where you want.<br>
<strong>Types of Position Values:</strong><br>
1) <strong>static (Default) : </strong><br>

> Elements are positioned according to the normal document flow.<br>

> No special positioning.

> Top, right, bottom, left properties have no effect.
relative

> Element remains in the normal flow.

> You can use top, right, bottom, left to move it relative to its normal position.

> Space is still reserved for the element in the layout.

2. <b>relative</b>

>Element remains in the normal flow.

>You can use top, right, bottom, left to move it relative to its normal position.

>Space is still reserved for the element in the layout.

3. <strong>absolute</strong>

>Element is removed from the normal flow.

>It is positioned relative to the nearest positioned ancestor (non-static).

>If no positioned ancestor, it uses the <html> element.

4. <strong>fixed</strong>

>Similar to absolute, but positioned relative to the viewport.

>Stays in the same place even when scrolling.
